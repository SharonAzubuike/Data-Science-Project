My next goal is to **visualize my data as a simple cluster using KMeans**

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
# Standardize the data
scaler = StandardScaler()
cardio_1_scaled = scaler.fit_transform(cardio_1)
```

```python
# Apply K-means clustering
num_clusters = 3  # I adjust the number of clusters based on my dataset
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
cardio_1['cluster'] = kmeans.fit_predict(cardio_1_scaled)
```

```python
pca = PCA(n_components=2)
cardio_1_pca = pca.fit_transform(cardio_1_scaled)

# Create a scatter plot of the clusters
plt.scatter(cardio_1_pca[:, 0], cardio_1_pca[:, 1], c=cardio_1['cluster'], cmap='viridis', alpha=0.5)
plt.title('K-means Clustering of Cardiovascular Dataset')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```


Bear in mind that Principal Component Analysis (PCA) is a dimensionality reduction technique that transforms the original features of a dataset into a new set of uncorrelated variables called principal components. Each principal component is a linear combination of the original features, and they are ordered by the amount of variance they explain in the data.
